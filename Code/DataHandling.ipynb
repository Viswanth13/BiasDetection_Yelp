```python
import pandas as pd

# Path to your mounted Drive file
reviews_json_path = '/content/drive/My Drive/yelp_academic_dataset_review.json'
output_csv_path   = 'review.csv'

# Choose a chunk size that fits comfortably in memory (e.g. 100k–500k rows)
chunk_size = 100_000

# Create an iterator over DataFrame chunks
json_chunks = pd.read_json(
    reviews_json_path,
    lines=True,
    chunksize=chunk_size
)

# Iterate and append to CSV
for i, chunk in enumerate(json_chunks):
    # On first chunk, write header; afterwards, append without header
    chunk.to_csv(
        output_csv_path,
        mode='w' if i == 0 else 'a',
        header=(i == 0),
        index=False
    )
    print(f'Processed chunk {i+1} — wrote {len(chunk)} rows')  # progress indicator
```

```python
import pandas as pd

# 3. Define file paths on Drive
biz_csv_path    = '/content/drive/My Drive/FairwAI/business.csv'
rev_csv_path    = '/content/drive/My Drive/FairwAI/review.csv'
output_dir      = '/content/drive/My Drive/FairwAI/EachCity/'
```

```python
biz_df = pd.read_csv(biz_csv_path, dtype={'business_id': str, 'categories': str})
rev_df = pd.read_csv(rev_csv_path, parse_dates=['date'], dtype={'business_id': str})
```

```python
biz_df.head()
```

```python
rev_df.head()
```

```python
# Filter for hospitality categories
hosp_terms = ['Restaurant','Hotel','Spa','Resort','Inn','Lodge','Bar','Cafe']
mask = biz_df['categories'].str.contains('|'.join(hosp_terms), na=False)
hosp_df = biz_df[mask].copy()
```

```python
# 6. Merge reviews with business info
merged = rev_df.merge(
    hosp_df[['business_id','name','categories','city','state',
             'latitude','longitude','review_count','stars']],
    on='business_id', how='inner'
)
```

```python
# target_cities = [
#     'Las Vegas', 'Toronto', 'Phoenix',
#     'Charlotte', 'Edinburgh', 'Pittsburgh'
# ]

target_cities = ['Philadelphia', 'Edmonton', 'Tampa','New Orleans', 'Indianapolis', 'Boise']
```

```python
# Sample top 12,000 recent reviews per city and write CSVs
for city in target_cities:
    city_df = (
        merged[merged['city'] == city]
        .sort_values('date', ascending=False)
        .groupby('city')
        .head(12000)
        .reset_index(drop=True)
    )
    # Ensure output directory exists
    !mkdir -p "{output_dir}"
    # Write per-city CSV
    filename = city.lower().replace(' ', '_') + '_reviews.csv'
    city_df.to_csv(f'{output_dir}{filename}', index=False)
    print(f'Wrote {len(city_df)} reviews for {city} to {filename}')
```
